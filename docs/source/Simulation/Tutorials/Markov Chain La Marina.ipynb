## Introduction
This Notebook explores Markov Chains using the example of the La Marina beach in Valencia, Spain. The goal is to model the number of people on each of the three stages of La Marina: EDEM (Education), Lanzadera (Acceleration), and Angels. Each stage is represented as a state in the Markov Chain, and the transitions between states are determined by the probabilities of people moving from one stage to another. The Notebook will cover how to set up the transition matrix, compute the steady-state distribution, and analyze the long-term behavior of the system.

This activity connects a physical, intuitive experiment in which we generated empirical data of the number of people in each stage of La Marina, with the theoretical framework of Markov Chains. By doing so, we can understand how to model real-world systems using Markov Chains and how to interpret the results in terms of the dynamics of the system.

```python
import pandas as pd

# Load sheets
edem = pd.read_csv("EDEM.csv")
lanzadera = pd.read_csv("Lanzadera.csv")
angels = pd.read_csv("Angels.csv")

# Add a column to identify the source of each transition
edem["From"] = "EDEM"
lanzadera["From"] = "Lanzadera"
angels["From"] = "Angels"

# Join the data into a single DataFrame
empirical_df = pd.concat([edem, lanzadera, angels])

# Reordenar columnas
empirical_df = empirical_df[["From", "To"]]

empirical_df.head()
```

## One-step Transition probability matrix
The following function computes the transition matrix from the data. It counts the number of transitions from each state to every other state and normalizes these counts to get probabilities.


```python
def compute_transition_matrix(data_frame):

    counts = pd.crosstab(data_frame['From'], data_frame['To'])
    P = counts.div(counts.sum(axis=1), axis=0)
    # reorder rows and columns in this order: EDEM, Lanzadera, Angels
    P = P.reindex(index=["EDEM", "Lanzadera", "Angels"], columns=["EDEM", "Lanzadera", "Angels"], fill_value=0)

    return P
```

## Steady-state distribution
Recall from class that the steady-state distribution of a Markov Chain is the distribution that remains unchanged after applying the transition matrix. It can be computed by solving the linear system of equations πP = π, where π is the steady-state distribution and P is the transition matrix, and forcing the condition that the sum of the probabilities in π is equal to 1.

## Bonus: Simulating the Markov Chain
To further analyze the behavior of the system, you can simulate the Markov Chain over a number of steps. This can be done by starting from an initial state and using the transition probabilities to determine the next state at each step. The following code simulates the Markov Chain for a given number of steps and records the state at each step.


```python
P = theoretical_transition_matrix.values  # Use the theoretical transition matrix for simulation
num_steps = 10
states = ["EDEM", "Lanzadera", "Angels"]
current_state = 0  # Start in EDEM
state_history = [states[current_state]]
for _ in range(num_steps):
    current_state = np.random.choice(len(states), p=P[current_state])
    state_history.append(states[current_state])

state_history
```

## Bonus: Alternative methods for computing the steady-state distribution
In addition to solving the linear system of equations, there are alternative methods for computing the steady-state distribution, such as using eigenvalues and eigenvectors. The steady-state distribution can be found as the eigenvector corresponding to the eigenvalue of 1 of the transition matrix. You can use the following code to compute the steady-state distribution using eigenvalues and eigenvectors:

```python
eigenvalues, eigenvectors = np.linalg.eig(theoretical_transition_matrix.values.T)
steady_state_vector = eigenvectors[:, np.isclose(eigenvalues, 1)]
steady_state_distribution = steady_state_vector / steady_state_vector.sum()
steady_state_distribution
```
